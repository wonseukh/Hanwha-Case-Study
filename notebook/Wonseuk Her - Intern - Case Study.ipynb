```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# 1.

```python
symbol_list = ['AAPL','SBUX','MSFT','CSCO','QCOM','FB','AMZN','TSLA','AMD','ZNGA']
data_list = []
  
for i in range(len(symbol_list)):
    temp_df = pd.read_csv("../data/" + symbol_list[i] + "_Historical_Data.csv")[::-1]
    temp_df.reset_index(drop=True, inplace=True)
    temp_df['Close/Last'] = temp_df['Close/Last'].str.strip('$').astype(float)
    data_list.append(temp_df[['Date','Close/Last']])
del temp_df
```

# 2.

#### Comment on the algorithm

There can be repeated Close prices within a span of 252 days.<br>
In that case, those repeated Close prices on separate dates, while having the same price, <br>
may end up with a different price momentum score.<br><br>
Therefore, I edited the algorithm so that the same price can receive the same score.<br>
With this edit, from an initial case where the score increases by 100/251 for every 1 increase in rank, <br>
a score can now increase by 100 / (251 - number of the repeated case).

```python
def momentum_calculator(df):
    momentum_list = []
    for i in range(252,len(df['Close/Last'])+1,1):
        temp = df['Close/Last'].iloc[i-252:i]
        temp_unique = temp.unique()
        temp_df = pd.DataFrame(temp_unique, columns=['Close']).sort_values(by="Close")
        temp_df.reset_index(drop=True, inplace=True)
        
        max_index = temp_df.idxmax()
        min_index = temp_df.idxmin()
        
        temp_df['Momentum'] = np.nan
        temp_df['Momentum'].iloc[max_index] = 100
        temp_df['Momentum'].iloc[min_index] = 0
        temp_df['Momentum'] = temp_df['Momentum'].interpolate()
        momentum_score = float(temp_df['Momentum'][temp_df['Close'] == df.iloc[i-1,1]])
        date = df['Date'].iloc[i-1]
        temp_list = [date,momentum_score]
        momentum_list.append(temp_list)
        
    df_momentum = pd.DataFrame(momentum_list,columns=['Date','Momentum'])
    df_momentum.set_index('Date',inplace=True)
    return df_momentum
```

```python
final_result = []
for i in range(0,len(data_list),1):
    temp = momentum_calculator(data_list[i])
    final_result.append(temp)

del temp
df_final = pd.concat(final_result,axis=1)
df_final.columns = symbol_list
```

# 3.

#### Comment on the algorithm

First of all, this algorithm operates in a shorter span of 20 days. <br>
Moreover, I have weighted those days so that the older Close prices are less represented and more recent prices are highly emphasized<br> 
By summing up these weighted prices, I have defined my expected price. 

The way I have scored is by comparing the Close price and expected price. When the Close price is below our expectation, the score will be  -1. <br>
If such a condition continues, the score will keep decreasing. <br>
However, as soon as the price exceeds our expectation, the score is reset, and a value of 1 is assigned. <br>
Again, if the price continues to exceed our expectations, the score will continuously increase until another inflection occurs. 

The strong suit of this algorithm is that it is based more on recent information compared to the previous algorithm. <br>
The previous algorithm made its inference from vast data, and, as a result, I would say my algorithm is quicker in catching a trend. 

Another strong point is its readability. The previous algorithm, while rich in content, can be hard to understand in one quick look.
On the other hand, my scores are in a recognizable range, and while the scores can be used for a whole picture, it can also be used to check specific ranges or periods. <br>

Graph below to illustrate:

```python
def my_momentum_calculator(df):
    momentum_list = []
    weight_list = []
    score_list = []
    for i in range(20,0,-1):
        weight_list.append(float((0.4*0.6**i)/0.5999780630493599))

    counter = 0
    for j in range(20,len(df['Close/Last'])+1,1):
        temp = df['Close/Last'].iloc[j-20:j]
        temp_list = []
        for k in range(0,20,1):
            temp_list.append(temp.iloc[k]*weight_list[k])
            
        close = temp.iloc[-1]
        weight_close = sum(temp_list)
        date = df['Date'].iloc[j-1]
        
        if(close > weight_close):
            counter= 1
        elif(close < weight_close):
            counter = -1
        score_list.append(counter)
        
        final_list = [date,close,weight_close]
        momentum_list.append(final_list)

    final_score = [score_list[0]]
    tracker = score_list[0]
    for L in range(1,len(score_list),1):
        if(score_list[L-1] == score_list[L]):
            tracker += score_list[L]
            final_score.append(tracker)
        else:
            tracker = 0
            tracker += score_list[L]
            final_score.append(tracker)
        
        
    df_momentum = pd.DataFrame(momentum_list,columns=['Date','Close','Weighted'])
    df_momentum.set_index('Date',inplace=True)
    df_momentum['Score'] = final_score
    return df_momentum
```

```python
final_result_v2 = []
for i in range(0,len(data_list),1):
    temp = my_momentum_calculator(data_list[i])['Score']
    final_result_v2.append(temp)

del temp
df_final_v2 = pd.concat(final_result_v2,axis=1)
df_final_v2.columns = symbol_list
```
